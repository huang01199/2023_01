```python
import matplotlib.pyplot as plt#說明書https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html

figure = plt.figure()#建立一個畫布
axes=figure.add_subplot(1,1,1,)#創建圖表大小
x=[0,1,2]
y=[3,4,5]
axes.plot(x,y)#只有線
axes.plot(x,y,"bo")#只有點
axes.plot(x,y,'bo--')#有虛線跟點
plt.show()
```

```python
import matplotlib.pyplot as plt

figure = plt.figure()#建立一個畫布
axes1 = figure.add_subplot(1,3,1)#第一列有三個，的第一個
axes1.set_title("title 1")
axes2 = figure.add_subplot(1,3,2)#第一列有三個，的第二個
axes2.set_title("title 2")
axes3 = figure.add_subplot(1,3,3)#第一列有三個，的第三個
axes3.set_title("title 3")

plt.show()
```

```python
figure1 = plt.figure()#建立一個畫布
axes=figure1.add_subplot(1,1,1,)#創建圖表大小
x=[0,1,2]
y=[3,4,5]
axes.plot(x,y)#只有線

figure2 = plt.figure()#建立第二個畫布
axes=figure2.add_subplot(1,1,1,)#創建圖表大小
x1=[2,3,6]
y1=[1,5,2]
axes.plot(x1,y1)#只有線

plt.show()
```
